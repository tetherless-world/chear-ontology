```python
def load_properties(url):
    properties = requests.get(properties_url).text
    properties = properties.split('\n')
    properties = [(x.split('=',maxsplit=1)) for x in properties if len(x) > 0]
    properties = dict([(key, value.split(',')) for key,value in properties])
    return properties

def make_namespaces(properties):
    return dict([(key,Namespace(value[0])) for key, value in properties.items()])

properties = load_properties(properties_url)
namespaces = make_namespaces(properties)

namespaces['ncit'] = Namespace('http://purl.obolibrary.org/obo/NCIT_')
namespaces['CSEO'] = Namespace('http://scai.fraunhofer.de/CSEO#')
```

```python
import numpy as np
from io import StringIO

sdd_files = [sdd_dir+f for f in os.listdir(sdd_dir) if 'SDD' in f and not f.startswith('~$')]

default_code_mappings = 'https://raw.githubusercontent.com/tetherless-world/chear-ontology/master/code_mappings.csv'

def replacenan(value):
    try:
        if np.isnan(value):
            return None
        else: return value
    except:
        return value

class SemanticDataDictionary(object):
    
    resource_columns = {
        'dictionary': ['Attribute','Unit','Entity','Role'],
        'codebook' : ['Class','Resource'],
        'timeline' : ['Type','Unit']
    }
    
    
    def _load_sheet(self, f, sheet, default=None):
        if sheet in self.infosheet and self.infosheet[sheet] is not None:
            dictionary_url = self.infosheet[sheet]
            if dictionary_url.startswith('#'):
                return pd.read_excel(f, dictionary_url[1:])
            else:
                return pd.read_csv(StringIO(requests.get(dictionary_url).text))
        elif default is not None:
            return pd.read_csv(StringIO(requests.get(default).text))
        
    def _map_uri(self, curie):
        if curie is None:
            return curie
        parts = curie.split(':',maxsplit=1)
        if parts[0] in self.namespaces:
            return self.namespaces[parts[0]][parts[1]]
        else:
            return curie
        
    def _map_uris(self, series):
        series = [replacenan(x) for x in series]
        series = [[x.strip() for x in y.split(',')] if y is not None else [] for y in series]
        series = [[self.code_mappings[replacenan(y)] if y in self.code_mappings else y for y in x]
                  for x in series]
        series = [[URIRef(self._map_uri(y)) for y in x] if x is not None else x for x in series]
        return series
    
    def _resolve_columns(self, sheet, columns):
        for col in columns:
            #print (self.filename, col)
            if col in sheet.columns:
                sheet[col] = self._map_uris(sheet[col])

    def __init__(self, f, namespaces):
        self.filename = f
        print (f)
        self.resources = set([])
        self.namespaces = namespaces
        self.infosheet = dict([(key, replacenan(value)) for key, value 
                               in pd.read_excel(f, 'InfoSheet',index_col=0).to_dict()["Value"].items()])
        self.code_mappings = self._load_sheet(f, 'Code_Mappings',default_code_mappings)
        self.code_mappings.index = self.code_mappings['code']
        self.code_mappings = self.code_mappings.to_dict()['uri']
        self.dictionary = self._load_sheet(f, 'Data_Dictionary')
        self.codebook = self._load_sheet(f, 'Codebook')
        self.timeline = self._load_sheet(f, 'Timeline')
        for attr, columns in self.resource_columns.items():
            sheet = self.__dict__[attr]
            self._resolve_columns(sheet, columns)
            for col in columns:
                if col in sheet.columns:
                    for cell in sheet[col]:
                        self.resources.update(cell)
                
sdds = [SemanticDataDictionary(sdd_file, namespaces) for sdd_file in sdd_files]
```
